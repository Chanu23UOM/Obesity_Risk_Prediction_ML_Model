```python
# OBESITY CLASSIFICATION - COMPLETE PIPELINE
# Enhanced version with data cleaning and ensemble methods
# Upload train.csv and test.csv to Colab before running

import os
import subprocess
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, ExtraTreesClassifier, VotingClassifier
from sklearn.metrics import accuracy_score
from imblearn.over_sampling import SMOTE
from xgboost import XGBClassifier
import lightgbm as lgb
from catboost import CatBoostClassifier

# ============================================
# HELPER FUNCTIONS
# ============================================
def has_nvidia_smi():
    """Check if NVIDIA GPU is available"""
    try:
        subprocess.run(['nvidia-smi'], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL, check=True)
        return True
    except Exception:
        return False

def normalize_yes_no(df):
    """Normalize all Yes/No variations across string columns"""
    for col in df.select_dtypes(include='object').columns:
        df[col] = (df[col]
                    .astype(str)
                    .str.strip()                # remove leading/trailing spaces
                    .str.lower()                # make lowercase
                    .replace({'yess': 'yes', 'ye': 'yes', 'y': 'yes',
                              'n': 'no', 'noo': 'no'}))
    return df

# ============================================
# 1. LOAD DATA
# ============================================
print("="*50)
print("üìÅ LOADING DATA")
print("="*50)
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
print(f"Train shape: {train.shape}, Test shape: {test.shape}")

# ============================================
# 2. DATA CLEANING
# ============================================
print("\n" + "="*50)
print("üßπ DATA CLEANING")
print("="*50)

# Normalize Yes/No variations
train = normalize_yes_no(train)
test = normalize_yes_no(test)
print("‚úÖ Normalized Yes/No variations")

# Sanitize column names (replace spaces -> underscores)
train.columns = train.columns.str.strip().str.replace(' ', '_').str.replace('-', '_')
test.columns = test.columns.str.strip().str.replace(' ', '_').str.replace('-', '_')
print("‚úÖ Sanitized column names")

# Keep PersonID for submission
if 'PersonID' in test.columns:
    person_ids = test['PersonID'].copy()
else:
    person_ids = pd.Series(range(1, len(test) + 1), name='PersonID')

# Drop PersonID from working dataframes
if 'PersonID' in train.columns:
    train = train.drop('PersonID', axis=1)
if 'PersonID' in test.columns:
    test = test.drop('PersonID', axis=1)

# Handle nulls and duplicates
print(f"Nulls in train: {train.isna().sum().sum()}")
train = train.drop_duplicates()
print(f"Duplicates removed: {train.duplicated().sum()}")

# Identify numeric and categorical columns
num_cols = train.select_dtypes(include="number").columns.tolist()
cat_cols = [c for c in train.select_dtypes(exclude="number").columns if c != 'Weight_Category']
print(f"\nNumerical columns ({len(num_cols)}): {num_cols[:5]}...")
print(f"Categorical columns ({len(cat_cols)}): {cat_cols}")

# ============================================
# 3. EXPLORATORY DATA ANALYSIS
# ============================================
print("\n" + "="*50)
print("üìä EXPLORATORY DATA ANALYSIS")
print("="*50)

print(f"\nTarget distribution:\n{train['Weight_Category'].value_counts()}")

# Correlation heatmap (optional - comment out if too slow)
try:
    plt.figure(figsize=(10, 8))
    corr_cols = [c for c in num_cols if c in train.columns]
    if 'Weight_Category' in train.columns:
        corr_cols.append('Weight_Category')
    corr = train[corr_cols].corr(numeric_only=True)
    sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', cbar=True)
    plt.title("Numeric Feature Correlation")
    plt.tight_layout()
    plt.show()
except Exception as e:
    print(f"‚ö†Ô∏è Plot skipped: {e}")

# ============================================
# 4. FEATURE ENGINEERING
# ============================================
print("\n" + "="*50)
print("üîß FEATURE ENGINEERING")
print("="*50)

df_train = train.copy()
df_test = test.copy()

# Convert boolean-like columns to integers
bool_cols = ['High_Calorie_Food', 'Family_History', 'Smoking_Habit']
for col in bool_cols:
    if col in df_train.columns:
        df_train[col] = (df_train[col].astype(str) == 'yes').astype(int)
    if col in df_test.columns:
        df_test[col] = (df_test[col].astype(str) == 'yes').astype(int)
print("‚úÖ Converted boolean columns")

# Label encode categorical columns
le = LabelEncoder()
cate_cols = ['Gender', 'Snack_Frequency', 'Alcohol_Consumption', 'Commute_Mode',
             'Physical_Activity_Level', 'Leisure_Time_Activity']
for col in cate_cols:
    if col in df_train.columns and col in df_test.columns:
        combined = pd.concat([df_train[col].astype(str), df_test[col].astype(str)], ignore_index=True)
        le.fit(combined)
        df_train[col] = le.transform(df_train[col].astype(str))
        df_test[col] = le.transform(df_test[col].astype(str))
print("‚úÖ Encoded categorical columns")

# Encode target variable
if 'Weight_Category' not in df_train.columns:
    raise KeyError("Train file must contain 'Weight_Category' column.")
le_target = LabelEncoder()
y_encoded = le_target.fit_transform(df_train['Weight_Category'])
print(f"‚úÖ Target encoded: {len(le_target.classes_)} classes")

# Prepare features
X_train = df_train.drop('Weight_Category', axis=1)
X_test = df_test.copy()

# Create BMI feature (CRITICAL for obesity classification)
if 'Weight_Kg' in X_train.columns and 'Height_cm' in X_train.columns:
    X_train['BMI'] = X_train['Weight_Kg'] / ((X_train['Height_cm'] / 100) ** 2)
    X_test['BMI'] = X_test['Weight_Kg'] / ((X_test['Height_cm'] / 100) ** 2)
    print("‚úÖ Created BMI feature")
else:
    print("‚ö†Ô∏è BMI columns not found; skipping BMI feature")

# Create Age-BMI interaction
if 'Age_Years' in X_train.columns and 'BMI' in X_train.columns:
    X_train['Age_BMI'] = X_train['Age_Years'] * X_train['BMI']
    X_test['Age_BMI'] = X_test['Age_Years'] * X_test['BMI']
    print("‚úÖ Created Age-BMI interaction")

# Additional features
if 'BMI' in X_train.columns:
    X_train['BMI_Squared'] = X_train['BMI'] ** 2
    X_test['BMI_Squared'] = X_test['BMI'] ** 2
    print("‚úÖ Created BMI squared feature")

if 'Activity_Level_Score' in X_train.columns and 'Screen_Time_Hours' in X_train.columns:
    X_train['Sedentary_Score'] = X_train['Screen_Time_Hours'] - X_train['Activity_Level_Score']
    X_test['Sedentary_Score'] = X_test['Screen_Time_Hours'] - X_test['Activity_Level_Score']
    print("‚úÖ Created Sedentary score")

if 'Activity_Level_Score' in X_train.columns and 'Water_Intake' in X_train.columns:
    X_train['Activity_Water'] = X_train['Activity_Level_Score'] * X_train['Water_Intake']
    X_test['Activity_Water'] = X_test['Activity_Level_Score'] * X_test['Water_Intake']
    print("‚úÖ Created Activity-Water interaction")

# Additional powerful interactions
if 'BMI' in X_train.columns and 'Family_Risk' in X_train.columns:
    X_train['BMI_Family_Risk'] = X_train['BMI'] * X_train['Family_Risk']
    X_test['BMI_Family_Risk'] = X_test['BMI'] * X_test['Family_Risk']
    print("‚úÖ Created BMI-Family Risk interaction")

if 'Activity_Level_Score' in X_train.columns and 'Vegetable_Intake' in X_train.columns:
    X_train['Healthy_Lifestyle'] = X_train['Activity_Level_Score'] * X_train['Vegetable_Intake']
    X_test['Healthy_Lifestyle'] = X_test['Activity_Level_Score'] * X_test['Vegetable_Intake']
    print("‚úÖ Created Healthy Lifestyle score")

if 'High_Calorie_Food' in X_train.columns and 'Water_Intake' in X_train.columns:
    X_train['Diet_Hydration'] = X_train['High_Calorie_Food'] / (X_train['Water_Intake'] + 0.1)
    X_test['Diet_Hydration'] = X_test['High_Calorie_Food'] / (X_test['Water_Intake'] + 0.1)
    print("‚úÖ Created Diet-Hydration ratio")

if 'Meal_Frequency' in X_train.columns and 'Water_Intake' in X_train.columns:
    X_train['Meal_Water_Balance'] = X_train['Meal_Frequency'] * X_train['Water_Intake']
    X_test['Meal_Water_Balance'] = X_test['Meal_Frequency'] * X_test['Water_Intake']
    print("‚úÖ Created Meal-Water balance")

print(f"\nFinal feature count: {X_train.shape[1]}")

# ============================================
# 5. CREATE TRAIN/HOLDOUT SPLIT (BEFORE ANY TRAINING)
# ============================================
print("\n" + "="*50)
print("üîç CREATING TRAIN/HOLDOUT SPLIT")
print("="*50)

# CRITICAL: Split BEFORE any model training to avoid leakage
X_temp, X_hold, y_temp, y_hold = train_test_split(
    X_train, y_encoded, test_size=0.2, stratify=y_encoded, random_state=42
)
print(f"Train set: {X_temp.shape}, Holdout set: {X_hold.shape}")

# ============================================
# 6. MODEL TRAINING
# ============================================
print("\n" + "="*50)
print("ü§ñ MODEL TRAINING")
print("="*50)

# Determine XGBoost tree method
tree_method = 'gpu_hist' if has_nvidia_smi() else 'hist'
print(f"Using XGBoost tree_method: {tree_method}")

# Train XGBoost on TRAINING data only
print("\nüöÄ Training XGBoost...")
xgb_model = XGBClassifier(
    objective='multi:softprob',
    num_class=len(le_target.classes_),
    random_state=42,
    n_estimators=500,
    max_depth=8,
    learning_rate=0.05,
    tree_method=tree_method,
    eval_metric='mlogloss'
)
xgb_model.fit(X_temp, y_temp)  # Train on X_temp, not X_train!
cv_xgb = cross_val_score(xgb_model, X_temp, y_temp, cv=StratifiedKFold(5, shuffle=True, random_state=42), scoring='accuracy')
xgb_hold_pred = xgb_model.predict(X_hold)
xgb_hold_acc = accuracy_score(y_hold, xgb_hold_pred)
print(f"‚úÖ XGB - CV: {cv_xgb.mean():.4f} | Holdout: {xgb_hold_acc:.4f}")

# Apply SMOTE only to training data (not holdout!)
print("\n‚öñÔ∏è Applying SMOTE to training data only...")
smote = SMOTE(random_state=42)
X_bal, y_bal = smote.fit_resample(X_temp, y_temp)
print(f"Original training class distribution: {np.bincount(y_temp)}")
print(f"Balanced training class distribution: {np.bincount(y_bal)}")

# ============================================
# 6. HOLDOUT VALIDATION (BEFORE SMOTE)
# ============================================
print("\n" + "="*50)
print("üîç CREATING HOLDOUT SET (BEFORE SMOTE)")
print("="*50)

# CRITICAL: Split BEFORE applying SMOTE to avoid leakage
X_temp, X_hold, y_temp, y_hold = train_test_split(
    X_train, y_encoded, test_size=0.2, stratify=y_encoded, random_state=42
)
print(f"Train set: {X_temp.shape}, Holdout set: {X_hold.shape}")

# Apply SMOTE only to training data (not holdout!)
print("\n‚öñÔ∏è Applying SMOTE to training data only...")
smote = SMOTE(random_state=42)
X_bal, y_bal = smote.fit_resample(X_temp, y_temp)
print(f"Original training class distribution: {np.bincount(y_temp)}")
print(f"Balanced training class distribution: {np.bincount(y_bal)}")

# ============================================
# 7. ENSEMBLE TRAINING
# ============================================
print("\nüéØ Training ensemble models...")
models = {
    'RandomForest': RandomForestClassifier(
        n_estimators=1000,
        max_depth=30,
        min_samples_split=2,
        min_samples_leaf=1,
        max_features='sqrt',
        class_weight='balanced',
        random_state=42,
        n_jobs=-1
    ),
    'GradientBoosting': GradientBoostingClassifier(
        n_estimators=300,
        learning_rate=0.05,
        max_depth=8,
        random_state=42
    ),
    'ExtraTrees': ExtraTreesClassifier(
        n_estimators=500,
        max_depth=20,
        min_samples_split=5,
        class_weight='balanced',
        random_state=42,
        n_jobs=-1
    ),
    'LightGBM': lgb.LGBMClassifier(
        n_estimators=500,
        max_depth=10,
        learning_rate=0.05,
        num_leaves=31,
        class_weight='balanced',
        random_state=42,
        verbose=-1
    ),
    'CatBoost': CatBoostClassifier(
        iterations=1000,
        learning_rate=0.05,
        depth=8,
        loss_function='MultiClass',
        random_state=42,
        verbose=0,
        thread_count=-1
    )
}

scores = {}
trained = {}
for name, model in models.items():
    try:
        print(f"\n  Training {name}...")
        cv = cross_val_score(model, X_bal, y_bal, cv=StratifiedKFold(5, shuffle=True, random_state=42), scoring='accuracy', n_jobs=-1)
        scores[name] = cv.mean()
        model.fit(X_bal, y_bal)
        trained[name] = model

        # Test on holdout (unseen data)
        hold_pred = model.predict(X_hold)
        hold_score = accuracy_score(y_hold, hold_pred)
        print(f"  ‚úÖ {name} - CV: {cv.mean():.4f} | Holdout: {hold_score:.4f}")
    except Exception as e:
        print(f"  ‚ö†Ô∏è {name} skipped due to error: {e}")

# Create weighted voting ensemble
print("\nüé≠ Creating Voting Ensemble...")
estimators = [(n, m) for n, m in trained.items()]
weights = [scores[n] for n, _ in estimators]
ensemble = VotingClassifier(estimators=estimators, voting='soft', weights=weights, n_jobs=-1)
ens_cv = cross_val_score(ensemble, X_bal, y_bal, cv=StratifiedKFold(5, shuffle=True, random_state=42), scoring='accuracy', n_jobs=-1)
ensemble.fit(X_bal, y_bal)
ens_hold_pred = ensemble.predict(X_hold)
ens_hold_acc = accuracy_score(y_hold, ens_hold_pred)
print(f"‚úÖ Ensemble - CV: {ens_cv.mean():.4f} | Holdout: {ens_hold_acc:.4f}")

# Select best model based on holdout performance
all_models = {
    'XGBoost': (xgb_model, cv_xgb.mean()),
    'Ensemble': (ensemble, ens_cv.mean())
}

# Get holdout scores for both
xgb_hold_pred = xgb_model.predict(X_hold)
xgb_hold_acc = accuracy_score(y_hold, xgb_hold_pred)

# ============================================
# 8. MODEL SELECTION & FINAL TRAINING
# ============================================
print("\n" + "="*50)
print("üìä MODEL SELECTION")
print("="*50)

print(f"\nüìä Final Comparison:")
print(f"XGBoost    - CV: {cv_xgb.mean():.4f} | Holdout: {xgb_hold_acc:.4f}")
print(f"Ensemble   - CV: {ens_cv.mean():.4f} | Holdout: {ens_hold_acc:.4f}")

# Select based on holdout performance (more reliable)
best_results = {
    'XGBoost': (xgb_hold_acc, cv_xgb.mean()),
    'RandomForest': (scores.get('RandomForest', 0), scores.get('RandomForest', 0)),
    'Ensemble': (ens_hold_acc, ens_cv.mean())
}

# Get holdout scores for all trained models
all_holdout_scores = {'XGBoost': xgb_hold_acc, 'Ensemble': ens_hold_acc}
for name, model in trained.items():
    hold_pred_temp = model.predict(X_hold)
    all_holdout_scores[name] = accuracy_score(y_hold, hold_pred_temp)
    print(f"{name:20} - Holdout: {all_holdout_scores[name]:.4f}")

# Select model with best holdout score
best_model_name = max(all_holdout_scores, key=all_holdout_scores.get)
hold_acc = all_holdout_scores[best_model_name]

if best_model_name == 'XGBoost':
    best_cv = cv_xgb.mean()
    print(f"\nüîÑ Retraining XGBoost on full training data...")
    best_model = XGBClassifier(
        objective='multi:softprob',
        num_class=len(le_target.classes_),
        random_state=42,
        n_estimators=500,
        max_depth=8,
        learning_rate=0.05,
        tree_method=tree_method,
        eval_metric='mlogloss'
    )
    best_model.fit(X_train, y_encoded)
elif best_model_name == 'Ensemble':
    best_cv = ens_cv.mean()
    print(f"\nüîÑ Retraining Ensemble on full training data...")
    best_model = VotingClassifier(estimators=estimators, voting='soft', weights=weights, n_jobs=-1)
    X_full_bal, y_full_bal = smote.fit_resample(X_train, y_encoded)
    best_model.fit(X_full_bal, y_full_bal)
else:
    # One of the individual models from ensemble
    best_cv = scores[best_model_name]
    print(f"\nüîÑ Retraining {best_model_name} on full training data...")
    X_full_bal, y_full_bal = smote.fit_resample(X_train, y_encoded)
    best_model = trained[best_model_name]
    best_model.fit(X_full_bal, y_full_bal)

print(f"\nüèÜ Best Model: {best_model_name}")
print(f"   CV Accuracy: {best_cv:.4f}")
print(f"   Holdout Accuracy: {hold_acc:.4f}")

# ============================================
# 9. PREDICTIONS & SUBMISSION
# ============================================
print("\n" + "="*50)
print("üìù GENERATING SUBMISSION")
print("="*50)

# Make predictions on test set
final_pred_encoded = best_model.predict(X_test)
final_y_pred = le_target.inverse_transform(final_pred_encoded)

# Create submission file
submission = pd.DataFrame({
    'PersonID': person_ids,
    'Weight_Category': final_y_pred
})
submission.to_csv('submission.csv', index=False)
print("‚úÖ Submission saved as 'submission.csv'")
print(f"\nSubmission preview:")
print(submission.head(10))
print(f"\nPrediction distribution:")
print(submission['Weight_Category'].value_counts())

# ============================================
# 10. FEATURE IMPORTANCE
# ============================================
print("\n" + "="*50)
print("üìä FEATURE IMPORTANCE")
print("="*50)

if hasattr(best_model, 'feature_importances_'):
    importances = pd.DataFrame({
        'Feature': X_train.columns,
        'Importance': best_model.feature_importances_
    }).sort_values('Importance', ascending=False)
    print("\nTop 15 features:")
    print(importances.head(15).to_string(index=False))

    # Plot feature importance
    try:
        plt.figure(figsize=(10, 6))
        importances.head(15).plot(kind='barh', x='Feature', y='Importance', legend=False)
        plt.xlabel('Importance')
        plt.title('Top 15 Feature Importances')
        plt.gca().invert_yaxis()
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"‚ö†Ô∏è Plot skipped: {e}")

# ============================================
# 11. SUMMARY
# ============================================
print("\n" + "="*50)
print("‚ú® PIPELINE COMPLETE")
print("="*50)
print(f"Best Model: {best_model_name}")
print(f"Cross-Validation Accuracy: {best_cv:.4f}")
print(f"Holdout Accuracy: {hold_acc:.4f}")
print(f"Submission file: submission.csv")
print("\nüöÄ Ready to submit to Kaggle!")
print("="*50)
```
